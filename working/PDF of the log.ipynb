PDF of the log

```python
import numpy as np
import random as rm
import matplotlib.pyplot as plt
import math
```

```python
N = 1000000
x = np.zeros(shape=N)

for i in range(N):
    x[i] = rm.uniform(0.1, 10.0)

plt.hist(x, bins=100, density = True, edgecolor='black', color='grey')

plt.xlabel('Samples')
plt.ylabel('Frequency')
plt.title('Uniform distribution histogram')

plt.show()
```

```python
y = np.zeros(shape=N)
y = np.log10(x)

y_sorted = np.sort(y)

plt.hist(y, bins=100, density = True, edgecolor='black', color='grey')

plt.xlabel('log(Samples)')
plt.ylabel('Frequency')
plt.title('Log of the uniform distribution histogram')

plt.show()
```

$$ p(y) = |\frac{d(log(x))}{dx}|^{-1}p(x) = |x|p(x) $$

```python
p_x = 1 / (10 - 0.1)
theo_y = np.zeros(shape=N)

for i in range(N):
    theo_y[i] = abs(10**(y_sorted[i])) * p_x

plt.hist(y_sorted, bins=100, density = True, edgecolor='black', color='grey')
#y_range = np.linspace(min(y_sorted), max(y_sorted), N)
plt.plot(y_sorted, theo_y, color = 'red')

plt.xlabel('$y$')
plt.ylabel('$pdf_{y}(y)$')
plt.title('Probability density function')

plt.show()
```

```python
log_mean_x = math.log10(np.mean(x))
mean_y = np.mean(y)

print('The logarithm of the mean of x is:' " " + str(log_mean_x))
print('The mean of y is:' " " + str(mean_y))

log_median_x = math.log10(np.median(x))
median_y = np.median(y)

print('The logarithm of the median of x is:' " " + str(log_median_x))
print('The median of y is:' " " + str(median_y))
```
