# PDF of the log

```python
import numpy as np
import random as rm
import matplotlib.pyplot as plt
import math
```

Use **numpy** to draw $N$ (where $N$ is large...) random samples from a uniform distribution between $0.1$ and $10$, and store these samples as $x$.

```python
N = 100000
x = np.random.uniform(0.1, 10, N)
```

Use matplotlib to make a histogram of these samples.

```python
plt.hist(x, bins=50, density = True, histtype='step', color='black')
plt.xlabel('Samples')
plt.ylabel('Frequency')
plt.title('Uniform distribution histogram')

plt.show()
```

Compute the base-10 log of your array $x$, and store this as $y$.

```python
y = np.log10(x)
y_sorted = np.sort(y)
```

Make another histogram for $y$. Using the equation to transform probability distributions, write what the theoretical pdf of $y$ is, and overplot it onto your histogram.

```python
plt.hist(y, bins=50, density=True, histtype='step', color='black')

plt.xlabel('log(Samples)')
plt.ylabel('Frequency')
plt.title('Log of the uniform distribution histogram')

plt.show()
```

$$ p(y) = |\frac{d(log(x))}{dx}|^{-1}p(x) = |x|p(x) = \frac {|10^{y}|}{10 - 0.1} $$

```python
p_x = 1 / (10 - 0.1)
theo_y = np.zeros(shape=N)

for i in range(N):
    theo_y[i] = abs(10**(y_sorted[i])) * np.log(10) * p_x
    
theo_y_sorted = np.sort(theo_y)

plt.hist(y_sorted, bins=50, density = True, histtype='step', color='black')
plt.plot(y_sorted, theo_y_sorted, color = 'red', linestyle='dashed')

plt.xlabel('$y$')
plt.ylabel('$pdf_{y}(y)$')
plt.title('Probability density function')

plt.show()
```

Compute the log of the mean of $x$ and the mean of $y$. Now compute the log of the median of $x$ and the median of $y$.

```python
log_mean_x = math.log10(np.mean(x))
mean_y = np.mean(y)

print("  The logarithm of the mean of x is", log_mean_x)
print("  The mean of y is", mean_y)

log_median_x = math.log10(np.median(x))
median_y = np.median(y)

print("  The logarithm of the median of x is", log_median_x)
print("  The median of y is", median_y)
```
