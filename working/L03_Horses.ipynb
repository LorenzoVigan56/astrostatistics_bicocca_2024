# Kicking horses

A famous early application of low-number statistics was an analysis of Prussian cavalryman horse-kick deaths by Bortkiewicz in 1898.

He studied the distribution of 122 men kicked to death by horses among 10 Prussian army corps within 20 years (so a total number of corpes of 200). He recorded the number of corps with a given number of deaths:  

**Number of deaths &emsp;&emsp; Number of groups**  
0 &emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&nbsp; 109  
1 &emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&nbsp; 65  
2 &emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&nbsp; 22  
3 &emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&nbsp; 3  
4 &emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;&nbsp; 1  

```python
import numpy as np
import random
import math
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.stats import poisson
import scipy.stats
```

I plot the resulting probability distribution.

```python
Dead_men = 122
tot_groups = 200

n_deaths = np.array([0, 1, 2, 3, 4])
n_groups = np.array([109, 65, 22, 3, 1])

total = np.sum(n_groups)

freq = n_groups/total

plt.scatter(n_deaths, freq, marker='.', color='black')
plt.title('Probability distribution')
plt.xlabel('Probability')
plt.ylabel('Number of deaths')
```

```python
plt.scatter(n_deaths, freq, marker='.', color='black')
plt.title('Probability distribution')
plt.xlabel('Probability')
plt.ylabel('Number of deaths')

for mu in [0.1, 0.5, 1, 2]:
    p = scipy.stats.poisson(mu).pmf(n_deaths)
    plt.plot(n_deaths, p, label='$\mu=' + str(mu)+'$')
plt.legend()
plt.show()
```

I try to fit it by hand.

```python
average_mu = np.average(n_deaths, weights = freq)
plt.scatter(n_deaths, freq, marker='.', color='black')
plt.title('Probability distribution')
plt.xlabel('Probability')
plt.ylabel('Number of deaths')

p = scipy.stats.poisson(average_mu).pmf(n_deaths)
plt.plot(n_deaths, p, color = 'red')
```
